```r
# =============================================================================
# CONFORMAL PREDICTION ANALYSIS: EUSTOCKMARKETS
# Enhanced with GARCH models and distribution analysis - Short Horizons
# =============================================================================

# Load required libraries
library(tidyverse)
library(ggplot2)
library(ahead)
library(forecast)
library(patchwork)
library(kableExtra)
library(viridis)
library(scales)
library(rugarch)       # For GARCH modeling
library(fGarch)        # For alternative GARCH implementations
library(moments)       # For skewness and kurtosis
library(goftest)       # For goodness-of-fit tests

# Configuration - SHORT HORIZONS
horizons <- c(1, 3, 5, 10)  # Updated to short-term horizons
min_train_size <- 500
LEVEL <- 99
GARCH_SPEC <- list(model = "sGARCH", garchOrder = c(1, 1), armaOrder = c(1, 1))

# Load EuStockMarkets data
cat("📊 Loading EuStockMarkets data...\n")
data("EuStockMarkets")
cat("✅ Data loaded successfully\n")
cat(sprintf("Dimensions: %d observations, %d stocks\n", nrow(EuStockMarkets), ncol(EuStockMarkets)))

# Prepare data for each stock
prepare_stock_data <- function(stock_prices, stock_name) {
  cat(sprintf("   Preparing %s: Converting prices to log returns...\n", stock_name))
  log_prices <- log(stock_prices)
  log_returns <- diff(log_prices)
  return(as.numeric(log_returns))
}

cat("🔄 Preparing stock data...\n")
stocks_data <- list(
  DAX = prepare_stock_data(EuStockMarkets[, "DAX"], "DAX"),
  SMI = prepare_stock_data(EuStockMarkets[, "SMI"], "SMI"),
  CAC = prepare_stock_data(EuStockMarkets[, "CAC"], "CAC"),
  FTSE = prepare_stock_data(EuStockMarkets[, "FTSE"], "FTSE")
)

# Data summary
cat("\n📈 Stock data summary (Log Returns):\n")
for(stock in names(stocks_data)) {
  returns <- stocks_data[[stock]]
  cat(sprintf("   %s: %d obs, mean=%.4f, sd=%.4f, skew=%.2f, kurt=%.2f\n",
              stock, length(returns), mean(returns, na.rm=TRUE),
              sd(returns, na.rm=TRUE), skewness(returns), kurtosis(returns)))
}

# Define methods - Added GARCH models
forecast_methods <- list(
  "rwf" = forecast::rwf,
  "naive" = forecast::naive,
  "snaive" = forecast::snaive,
  "auto.arima" = forecast::auto.arima,
  "ets" = forecast::ets,
  "tbats" = forecast::tbats,
  "nnetar" = forecast::nnetar,
  "garch_rugarch" = function(y, h, level) {
    # Placeholder - will be handled separately
    return(NULL)
  },
  "garch_fgarch" = function(y, h, level) {
    # Placeholder - will be handled separately
    return(NULL)
  }
)

conformal_methods <- c("block-bootstrap", "surrogate", "kde", "bootstrap", "fitdistr")

cat(sprintf("\n🔧 Configured %d forecast methods and %d conformal methods\n",
            length(forecast_methods), length(conformal_methods)))

# GARCH forecasting function
forecast_garch <- function(train_data, h, level, package = "rugarch") {
  tryCatch({
    alpha <- (100 - level)/100

    if(package == "rugarch") {
      # rugarch implementation
      spec <- ugarchspec(variance.model = list(model = "sGARCH", garchOrder = c(1,1)),
                        mean.model = list(armaOrder = c(1,1), include.mean = TRUE),
                        distribution.model = "std")

      fit <- ugarchfit(spec, train_data)
      forecast <- ugarchforecast(fit, n.ahead = h)

      mean_fcst <- as.numeric(forecast@forecast$seriesFor)
      sigma_fcst <- as.numeric(forecast@forecast$sigmaFor)
      df <- coef(fit)["shape"]

      # Student's t distribution intervals
      lower <- mean_fcst + sigma_fcst * qt(alpha/2, df = df)
      upper <- mean_fcst + sigma_fcst * qt(1 - alpha/2, df = df)

    } else if(package == "fgarch") {
      # fGarch implementation
      fit <- garchFit(~ arma(1,1) + garch(1,1), data = train_data, cond.dist = "std", trace = FALSE)
      forecast <- predict(fit, n.ahead = h)

      mean_fcst <- forecast$meanForecast
      sigma_fcst <- forecast$standardDeviation
      df <- coef(fit)["shape"]

      lower <- mean_fcst + sigma_fcst * qt(alpha/2, df = df)
      upper <- mean_fcst + sigma_fcst * qt(1 - alpha/2, df = df)
    }

    # Create forecast object
    result <- list(
      mean = mean_fcst,
      lower = matrix(lower, ncol = 1),
      upper = matrix(upper, ncol = 1)
    )
    class(result) <- "forecast"
    return(result)

  }, error = function(e) {
    return(NULL)
  })
}

# Enhanced helper functions for GARCH support
calculate_coverage <- function(forecast_obj, actual_values, level = 99) {
  if(is.null(forecast_obj)) return(NA)

  if(inherits(forecast_obj, "forecast") && !is.null(forecast_obj$lower)) {
    # Handle standard forecast objects
    lower_bounds <- as.numeric(forecast_obj$lower)
    upper_bounds <- as.numeric(forecast_obj$upper)
  } else {
    # Handle ahead::conformalize objects
    lower_col <- paste0("Lo ", level)
    upper_col <- paste0("Hi ", level)
    forecast_df <- as.data.frame(forecast_obj)

    if(!(lower_col %in% colnames(forecast_df)) || !(upper_col %in% colnames(forecast_df))) {
      return(NA)
    }

    lower_bounds <- forecast_df[, lower_col]
    upper_bounds <- forecast_df[, upper_col]
  }

  n_test <- min(length(actual_values), length(lower_bounds))
  if(n_test == 0) return(NA)

  covered <- sum(actual_values[1:n_test] >= lower_bounds[1:n_test] &
                 actual_values[1:n_test] <= upper_bounds[1:n_test], na.rm = TRUE)

  return(covered / n_test)
}

calculate_winkler <- function(forecast_obj, actual_values, level = 99) {
  if(is.null(forecast_obj)) return(NA)

  if(inherits(forecast_obj, "forecast") && !is.null(forecast_obj$lower)) {
    lower_bounds <- as.numeric(forecast_obj$lower)
    upper_bounds <- as.numeric(forecast_obj$upper)
  } else {
    lower_col <- paste0("Lo ", level)
    upper_col <- paste0("Hi ", level)
    forecast_df <- as.data.frame(forecast_obj)

    if(!(lower_col %in% colnames(forecast_df)) || !(upper_col %in% colnames(forecast_df))) {
      return(NA)
    }

    lower_bounds <- forecast_df[, lower_col]
    upper_bounds <- forecast_df[, upper_col]
  }

  alpha <- (100 - level) / 100
  n_test <- min(length(actual_values), length(lower_bounds))
  if(n_test == 0) return(NA)

  widths <- upper_bounds[1:n_test] - lower_bounds[1:n_test]

  penalties <- ifelse(actual_values[1:n_test] < lower_bounds[1:n_test],
                     (2/alpha) * (lower_bounds[1:n_test] - actual_values[1:n_test]),
                     ifelse(actual_values[1:n_test] > upper_bounds[1:n_test],
                           (2/alpha) * (actual_values[1:n_test] - upper_bounds[1:n_test]),
                           0))

  return(mean(widths + penalties, na.rm = TRUE))
}

# Distribution analysis function
analyze_distribution <- function(forecast_obj, actual_values, level = 99) {
  if(is.null(forecast_obj)) return(list(skewness = NA, kurtosis = NA, ad_test = NA, ks_test = NA))

  if(inherits(forecast_obj, "forecast") && !is.null(forecast_obj$lower)) {
    lower_bounds <- as.numeric(forecast_obj$lower)
    upper_bounds <- as.numeric(forecast_obj$upper)
    means <- as.numeric(forecast_obj$mean)
  } else {
    lower_col <- paste0("Lo ", level)
    upper_col <- paste0("Hi ", level)
    forecast_df <- as.data.frame(forecast_obj)

    if(!(lower_col %in% colnames(forecast_df)) || !(upper_col %in% colnames(forecast_df))) {
      return(list(skewness = NA, kurtosis = NA, ad_test = NA, ks_test = NA))
    }

    lower_bounds <- forecast_df[, lower_col]
    upper_bounds <- forecast_df[, upper_col]
    means <- forecast_df$`Point Forecast`
  }

  n_test <- min(length(actual_values), length(means))
  if(n_test == 0) return(list(skewness = NA, kurtosis = NA, ad_test = NA, ks_test = NA))

  # Calculate standardized residuals
  residuals <- (actual_values[1:n_test] - means[1:n_test]) / ((upper_bounds[1:n_test] - lower_bounds[1:n_test])/ (2 * qnorm(1 - (100-level)/200)))

  # Remove infinite values
  residuals <- residuals[is.finite(residuals)]

  if(length(residuals) < 5) return(list(skewness = NA, kurtosis = NA, ad_test = NA, ks_test = NA))

  # Distribution metrics
  skew <- skewness(residuals)
  kurt <- kurtosis(residuals)

  # Goodness-of-fit tests
  tryCatch({
    ad_test <- ad.test(residuals, "pnorm")$p.value
    ks_test <- ks.test(residuals, "pnorm")$p.value
  }, error = function(e) {
    ad_test <- NA
    ks_test <- NA
  })

  return(list(
    skewness = skew,
    kurtosis = kurt,
    ad_test = ad_test,
    ks_test = ks_test
  ))
}

# Main analysis
run_conformal_analysis <- function() {
  cat("\n🎯 === STARTING ENHANCED ANALYSIS - SHORT HORIZONS ===\n")

  all_results <- data.frame()
  total_combinations <- (length(stocks_data) * length(horizons) *
                         (length(forecast_methods) - 2) * length(conformal_methods)) + # Standard methods
                        (length(stocks_data) * length(horizons) * 2) # GARCH methods
  current <- 0

  cat(sprintf("Total combinations to test: %d\n", total_combinations))

  for(stock_name in names(stocks_data)) {
    cat(sprintf("\n📊 Processing %s (%d returns)...\n", stock_name, length(stocks_data[[stock_name]])))

    stock_returns <- stocks_data[[stock_name]]

    for(h in horizons) {
      cat(sprintf("   ⏰ Horizon: %d days\n", h))

      n_total <- length(stock_returns)
      n_train <- n_total - h - 20  # Reduced calibration set for shorter horizons

      if(n_train < min_train_size) {
        cat(sprintf("   ⚠️  Skipping - training size (%d) < minimum (%d)\n", n_train, min_train_size))
        next
      }

      train_data <- stock_returns[1:n_train]
      test_data <- stock_returns[(n_train + 1):(n_train + h)]
      train_ts <- ts(train_data)

      cat(sprintf("   📁 Train: %d obs, Test: %d obs\n", length(train_data), length(test_data)))

      # Process standard methods
      for(method_name in names(forecast_methods)) {
        if(method_name %in% c("garch_rugarch", "garch_fgarch")) next # Skip GARCH here

        cat(sprintf("   🔍 %s: ", method_name))

        for(conf_method in conformal_methods) {
          current <- current + 1

          if(current %% 10 == 0) {
            cat(sprintf("\n      Progress: %.1f%%", (current/total_combinations)*100))
          }

          tryCatch({
            conf_forecast <- ahead::conformalize(
              FUN = forecast_methods[[method_name]],
              y = train_ts,
              h = h,
              level = LEVEL,
              method = conf_method
            )

            coverage <- calculate_coverage(conf_forecast, test_data, LEVEL)
            winkler <- calculate_winkler(conf_forecast, test_data, LEVEL)
            dist_analysis <- analyze_distribution(conf_forecast, test_data, LEVEL)

            forecast_df <- as.data.frame(conf_forecast)
            lower_col <- paste0("Lo ", LEVEL)
            upper_col <- paste0("Hi ", LEVEL)

            avg_width <- if(lower_col %in% colnames(forecast_df) &&
                           upper_col %in% colnames(forecast_df)) {
              mean(forecast_df[, upper_col] - forecast_df[, lower_col], na.rm = TRUE)
            } else {
              NA
            }

            result_row <- data.frame(
              stock = stock_name,
              horizon = h,
              forecast_method = method_name,
              conformal_method = conf_method,
              coverage_rate = coverage,
              winkler_score = winkler,
              avg_width = avg_width,
              target_coverage = LEVEL/100,
              coverage_deviation = abs(coverage - LEVEL/100),
              resid_skewness = dist_analysis$skewness,
              resid_kurtosis = dist_analysis$kurtosis,
              ad_test_pvalue = dist_analysis$ad_test,
              ks_test_pvalue = dist_analysis$ks_test,
              stringsAsFactors = FALSE
            )

            all_results <- rbind(all_results, result_row)
            cat("✅")

          }, error = function(e) {
            cat("❌")
          })
        }
        cat("\n")
      }

      # Process GARCH methods separately
      cat("   🔍 GARCH methods: ")
      for(garch_type in c("rugarch", "fgarch")) {
        current <- current + 1

        tryCatch({
          garch_forecast <- forecast_garch(train_data, h, LEVEL, garch_type)

          coverage <- calculate_coverage(garch_forecast, test_data, LEVEL)
          winkler <- calculate_winkler(garch_forecast, test_data, LEVEL)
          dist_analysis <- analyze_distribution(garch_forecast, test_data, LEVEL)

          avg_width <- if(!is.null(garch_forecast$lower) && !is.null(garch_forecast$upper)) {
            mean(garch_forecast$upper - garch_forecast$lower, na.rm = TRUE)
          } else {
            NA
          }

          result_row <- data.frame(
            stock = stock_name,
            horizon = h,
            forecast_method = paste0("garch_", garch_type),
            conformal_method = "garch",
            coverage_rate = coverage,
            winkler_score = winkler,
            avg_width = avg_width,
            target_coverage = LEVEL/100,
            coverage_deviation = abs(coverage - LEVEL/100),
            resid_skewness = dist_analysis$skewness,
            resid_kurtosis = dist_analysis$kurtosis,
            ad_test_pvalue = dist_analysis$ad_test,
            ks_test_pvalue = dist_analysis$ks_test,
            stringsAsFactors = FALSE
          )

          all_results <- rbind(all_results, result_row)
          cat("✅")

        }, error = function(e) {
          cat("❌")
        })
      }
      cat("\n")
    }
  }

  cat(sprintf("\n🎉 Completed! Generated %d results.\n", nrow(all_results)))
  return(all_results)
}

# Run analysis
results <- run_conformal_analysis()

# Clean results
results_clean <- results %>%
  filter(!is.na(coverage_rate) & !is.na(winkler_score) &
         !is.infinite(coverage_rate) & !is.infinite(winkler_score))

cat(sprintf("🧹 Clean results: %d rows (%.1f%% success rate)\n",
            nrow(results_clean), (nrow(results_clean)/nrow(results))*100))

# Create leaderboards PER STOCK
create_stock_leaderboards <- function(data, stock_name, horizon_val) {
  stock_data <- data %>% filter(stock == stock_name, horizon == horizon_val)

  if(nrow(stock_data) == 0) return(NULL)

  # Coverage leaderboard
  coverage_board <- stock_data %>%
    arrange(coverage_deviation, desc(coverage_rate)) %>%
    mutate(
      rank = row_number(),
      coverage_pct = sprintf("%.1f%%", coverage_rate * 100),
      deviation_pct = sprintf("%.1f%%", coverage_deviation * 100),
      method_combo = paste(forecast_method, conformal_method, sep = " + "),
      width_rounded = round(avg_width, 6)
    ) %>%
    select(rank, method_combo, coverage_pct, deviation_pct, width_rounded) %>%
    head(10)

  # Winkler leaderboard
  winkler_board <- stock_data %>%
    arrange(winkler_score) %>%
    mutate(
      rank = row_number(),
      coverage_pct = sprintf("%.1f%%", coverage_rate * 100),
      winkler_rounded = round(winkler_score, 6),
      method_combo = paste(forecast_method, conformal_method, sep = " + "),
      width_rounded = round(avg_width, 6)
    ) %>%
    select(rank, method_combo, winkler_rounded, coverage_pct, width_rounded) %>%
    head(10)

  return(list(coverage = coverage_board, winkler = winkler_board))
}

# Display leaderboards PER STOCK AND HORIZON
cat("\n🏆 === LEADERBOARDS BY STOCK AND HORIZON ===\n")

for(stock in names(stocks_data)) {
  cat(sprintf("\n⭐ %s - TOP PERFORMERS BY HORIZON ⭐\n", stock))

  for(h in horizons) {
    cat(sprintf("\n   📅 %d-DAY HORIZON\n", h))

    boards <- create_stock_leaderboards(results_clean, stock, h)

    if(!is.null(boards)) {
      cat("\n   TOP 10 - COVERAGE RATE (Closest to 99%)\n")
      print(kable(boards$coverage, format = "simple",
                  caption = paste(stock, "-", h, "Day Horizon - Coverage")))

      cat("\n   TOP 10 - WINKLER SCORE (Most Efficient)\n")
      print(kable(boards$winkler, format = "simple",
                  caption = paste(stock, "-", h, "Day Horizon - Efficiency")))
    }
  }
}

# Create plots PER STOCK
create_stock_plots <- function(data, stock_name) {
  stock_data <- data %>% filter(stock == stock_name)

  if(nrow(stock_data) == 0) return(NULL)

  # Plot 1: Coverage by conformal method (PER STOCK)
  p1 <- stock_data %>%
    ggplot(aes(x = conformal_method, y = coverage_rate, fill = conformal_method)) +
    geom_boxplot(alpha = 0.7) +
    geom_hline(yintercept = LEVEL/100, color = "red", linetype = "dashed", linewidth = 1) +
    facet_wrap(~ horizon, labeller = label_both) +
    scale_fill_viridis_d() +
    scale_y_continuous(labels = percent_format()) +
    labs(
      title = paste("Coverage Rate for", stock_name),
      subtitle = paste("Target:", LEVEL, "% (red line)"),
      x = "Conformal Method", y = "Coverage Rate", fill = "Method"
    ) +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1))

  # Plot 2: Winkler scores by forecast method (using log transformation)
  p2 <- stock_data %>%
    mutate(log_winkler = log1p(winkler_score)) %>%
    ggplot(aes(x = forecast_method, y = log_winkler, fill = forecast_method)) +
    geom_boxplot(alpha = 0.7) +
    facet_wrap(~ horizon) +
    scale_fill_viridis_d(option = "plasma") +
    labs(
      title = paste("Log(Winkler Score + 1) for", stock_name),
      subtitle = "Lower = better performance (log scale for better visualization)",
      x = "Forecast Method", y = "Log(1 + Winkler Score)", fill = "Method"
    ) +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1))

  # Plot 3: Coverage vs Winkler tradeoff (PER HORIZON and PER STOCK)
  p3_list <- list()
  for(h in horizons) {
    horizon_data <- stock_data %>% filter(horizon == h)
    if(nrow(horizon_data) > 0) {
      p3_list[[as.character(h)]] <- horizon_data %>%
        ggplot(aes(x = coverage_deviation, y = winkler_score,
                  color = forecast_method, shape = conformal_method)) +
        geom_point(size = 3, alpha = 0.8) +
        scale_color_viridis_d(option = "turbo") +
        labs(
          title = paste(stock_name, "-", h, "Day Horizon: Coverage vs Winkler"),
          subtitle = "Best methods: bottom-left (accurate and efficient)",
          x = "Coverage Deviation", y = "Winkler Score",
          color = "Forecast Method", shape = "Conformal Method"
        ) +
        theme_minimal() +
        theme(legend.position = "bottom")
    }
  }

  # Plot 4: Distribution analysis
  p4 <- stock_data %>%
    filter(!is.na(resid_skewness) & !is.na(resid_kurtosis)) %>%
    ggplot(aes(x = resid_skewness, y = resid_kurtosis,
               color = forecast_method, size = coverage_deviation)) +
    geom_point(alpha = 0.7) +
    geom_hline(yintercept = 0, linetype = "dashed") +
    geom_vline(xintercept = 0, linetype = "dashed") +
    facet_wrap(~ horizon) +
    scale_color_viridis_d(option = "inferno") +
    labs(
      title = paste("Residual Distribution for", stock_name),
      subtitle = "Skewness and Kurtosis of standardized residuals",
      x = "Skewness", y = "Kurtosis",
      color = "Forecast Method", size = "Coverage Deviation"
    ) +
    theme_minimal()

  return(list(p1 = p1, p2 = p2, p3 = p3_list, p4 = p4))
}

# Generate and display plots PER STOCK
cat("\n📊 === GENERATING PLOTS PER STOCK ===\n")

for(stock in names(stocks_data)) {
  cat(sprintf("\n🖼️  Creating plots for %s...\n", stock))
  stock_plots <- create_stock_plots(results_clean, stock)

  if(!is.null(stock_plots)) {
    print(stock_plots$p1)
    print(stock_plots$p2)

    # Print each horizon-specific plot
    for(h_name in names(stock_plots$p3)) {
      print(stock_plots$p3[[h_name]])
    }

    print(stock_plots$p4)
    cat(sprintf("✅ Plots completed for %s\n", stock))
  }
}

# Best overall performers PER STOCK AND HORIZON
cat("\n🏅 === BEST PERFORMERS BY STOCK AND HORIZON ===\n")

for(stock in names(stocks_data)) {
  cat(sprintf("\n⭐ %s - BEST METHODS BY HORIZON ⭐\n", stock))

  for(h in horizons) {
    cat(sprintf("\n   📅 %d-DAY HORIZON\n", h))

    best_horizon <- results_clean %>%
      filter(stock == !!stock, horizon == h) %>%
      arrange(coverage_deviation, winkler_score) %>%
      head(5) %>%
      mutate(
        method_combo = paste(forecast_method, conformal_method, sep = " + "),
        coverage_pct = sprintf("%.1f%%", coverage_rate * 100),
        winkler_rounded = round(winkler_score, 6),
        width_rounded = round(avg_width, 6),
        skew_rounded = round(resid_skewness, 2),
        kurt_rounded = round(resid_kurtosis, 2)
      )

    if(nrow(best_horizon) > 0) {
      print(kable(best_horizon %>%
                  select(method_combo, coverage_pct, winkler_rounded,
                         width_rounded, skew_rounded, kurt_rounded),
                caption = paste("Best Performers for", stock, "-", h, "Day Horizon"),
                col.names = c("Method", "Coverage", "Winkler", "Width", "Skew", "Kurt")))
    }
  }
}

# Distribution analysis summary
cat("\n📊 === DISTRIBUTION ANALYSIS SUMMARY ===\n")

dist_summary <- results_clean %>%
  group_by(stock, forecast_method) %>%
  summarise(
    avg_skewness = mean(resid_skewness, na.rm = TRUE),
    avg_kurtosis = mean(resid_kurtosis, na.rm = TRUE),
    norm_reject_rate = mean(ad_test_pvalue < 0.05 | ks_test_pvalue < 0.05, na.rm = TRUE),
    n_obs = n(),
    .groups = "drop"
  ) %>%
  filter(n_obs > 3)

print(kable(dist_summary, digits = 3,
            caption = "Average Residual Distribution Characteristics by Method"))

# GARCH performance comparison
cat("\n⚡ === GARCH MODELS PERFORMANCE ===\n")

garch_performance <- results_clean %>%
  filter(grepl("garch", forecast_method)) %>%
  group_by(stock, forecast_method, horizon) %>%
  summarise(
    avg_coverage = mean(coverage_rate, na.rm = TRUE),
    avg_winkler = mean(winkler_score, na.rm = TRUE),
    avg_width = mean(avg_width, na.rm = TRUE),
    .groups = "drop"
  )

print(kable(garch_performance, digits = 4,
            caption = "GARCH Models Performance Summary"))

# Final summary
cat("\n🎯 === ANALYSIS COMPLETE ===\n")
cat(sprintf("Stocks analyzed: %s\n", paste(names(stocks_data), collapse = ", ")))
cat(sprintf("Horizons: %s days\n", paste(horizons, collapse = ", ")))
cat(sprintf("Total successful results: %d\n", nrow(results_clean)))
cat(sprintf("GARCH models included: rugarch, fGarch\n"))
cat(sprintf("Distribution metrics calculated: skewness, kurtosis, normality tests\n"))

cat("\n✅ Enhanced analysis with short horizons completed successfully!\n")
```
